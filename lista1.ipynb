## Zadanie 2 (15pt)

Napisz system rekomendacji filmów. Systemy takie są wykorzystywane przez różne firmy np.
Netflix organizował konkurs na opracowanie algorytmu, który będzie przewidywał
ocenę użytkownika [Netflix Prize](https://en.wikipedia.org/wiki/Netflix_Prize).
W zadaniu tym zaimplementujemy podobny system, który jednak zamiast przewidywania będzie na
podstawie preferencji użytkownika rekomendował filmy, które najprawdopodobniej mu się spodobają.
Istnieje wiele sposobów, aby taki system napisać, dla zainteresowanych bardziej tematem proponuje
zobaczyć np. [Recommendation Systems](http://infolab.stanford.edu/~ullman/mmds/ch9.pdf).
W tym zadaniu wybierzemy w miarę prosty i łatwy do implementacji system rekomendacji.
Sformalizujmy problem. Załóżmy, że mamy macierz oceny, gdzie wiersze będą reprezentować
użytkowników a kolumny filmy np.
$$ \begin{array}{ccc}
    & \mbox{Matrix} & \mbox{Star Wars IV} \\
    \mbox{Alice} & 5 & 4 \\
    \mbox{Bob} & 0 & 1 \\
    \mbox{John} & 2 & 2 \\
    \mbox{Ada} & 5 & 5
   \end{array}$$
Patrząc na powyższą macierz widać, że kolumny pierwsza i druga mają podobne oceny stąd można wywnioskować, że
filmy Matrix i Star Wars IV są (według użytkowników) podobne do siebie, czyli jeśli komuś podobał się Matrix
to jest duża szansa, że będzie podobał mu się Star Wars i odwrotnie. Dlatego "podobieństwo" sformalizujemy przez wykorzystanie
[podobieństwa cosinusowego](https://en.wikipedia.org/wiki/Cosine_similarity), czyli
jeśli $x = (x_1, x_2, \dots, x_n)$ i $y = (y_1, y_2, \dots, y_n)$, wtedy:
$$x_1 y_1 + x_2 y_2 + \cdots + x_n y_n = x \cdot y = \lVert x \rVert \lVert y \rVert \cos(\theta)$$
Gdzie $\theta$ jest kątem między wektorami. Wygodniej będzie nam normalizować wektory. Wtedy
$$x_1 y_1 + x_2 y_2 + \cdots + x_n y_n = \cos(\theta).$$ Wtedy $\cos(\theta)$ reprezentuje
podobieństwo jednego wektora do drugiego, czyli jak bardzo jedne oceny są bliskie innym.

Dla naszej macierzy otrzymujemy

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
 
 ####################################### part 1 #######################################
# Wczytanie danych z pliku ratings.csv
ratings_df = pd.read_csv("ratings.csv")

# Wybieramy oceny użytkowników dla filmu 'Toy Story (1995)' (movieId=1)
toy_story_ratings = ratings_df[ratings_df['movieId'] == 1]

# Lista identyfikatorów użytkowników, którzy ocenili 'Toy Story'
toy_story_user_ids = toy_story_ratings['userId'].tolist()
#print(len(toy_story_user_ids))
#print(toy_story_user_ids)

# Wybieramy tylko te wiersze z ocenami, które odpowiadają użytkownikom oceniającym 'Toy Story'
selected_ratings = ratings_df[ratings_df['userId'].isin(toy_story_user_ids)]

# Tworzymy macierz ocen (X) i wektor ocen 'Toy Story' (y)
X = selected_ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)

X = X.drop(columns=[1])

y = selected_ratings[selected_ratings['movieId'] == 1].set_index('userId')['rating']
print(X.head(10))
#print(y.head(10))

# Lista różnych wartości m (liczba kolumn)
m_values = [10, 100, 1000, 10000]

# Lista, do której będziemy zapisywać błędy dla różnych wartości m
errors = []

# caly zbior z ocenami
X_all = ratings_df.pivot(index='userId', columns='movieId', values='rating').fillna(0)
X_all = X_all.drop(columns=[1])

"""
Wykonujemy regresje 
"""

# Dla każdej wartości m wykonujemy regresję liniową i obliczamy błąd
# Lista, do której będziemy zapisywać wartości współczynnika determinacji (R^2) dla różnych wartości m
r_squared_values = []
for m in m_values:
    # Wybieramy m pierwszych filmów
    first_movies = X.columns[:m]
    # Tworzymy X zawierające tylko m pierwszych filmów
    X_subset = X[first_movies]
    
    X_all_subset = X_all[first_movies]
    # Tworzymy model regresji liniowej
    model = LinearRegression()
    # Dopasowujemy model do danych, trenujemy model
    model.fit(X_subset, y)
    # Przewidujemy oceny 'Toy Story' dla WSZYSTKICH użytkowników
    y_pred = model.predict(X_all_subset)
    #print(len(y_pred))
    # ze zbioru predycji wybieramy tylko uzytkownikow ktorzy oceniali toy story
    y_pred_selected = y_pred[X_all_subset.index.isin(toy_story_user_ids)]
    
    #print("y_pred_selected: ", y_pred_selected)
    
    # Obliczamy błąd (MSE) dla przewidywanych ocen
    error = mean_squared_error(y, y_pred_selected)
    errors.append(error)
    #print("Errors: ",errors)
    
    # Obliczamy współczynnik determinacji (R^2) dla WSZYSTKICH użytkowników
    r_squared = r2_score(y, y_pred_selected)
    r_squared_values.append(r_squared)

# Wykres zależności błędu od liczby kolumn (wartości m)
plt.plot(m_values, errors, marker='o')
plt.title('Zależność błędu od liczby kolumn (wartości m) dla zbioru treningowego')
plt.xlabel('Liczba kolumn (m)')
plt.ylabel('Błąd (MSE)')
plt.grid(True)
plt.show()

plt.plot(m_values, r_squared_values, marker='o')
plt.title('Zależność współczynnika determinacji od liczby kolumn (wartości m) dla zbioru treningowego')
plt.xlabel('Liczba kolumn (m)')
plt.ylabel('Współczynnik determinacji (R^2)')
plt.grid(True)
plt.xscale('log')  # Ustawienie skali logarytmicznej na osi poziomej
plt.show()

####################################### part 2 #######################################

# Podział zbioru osób na zbiór treningowy i zbiór testowy
train_size = 200
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

# Lista różnych wartości m (liczba kolumn)
m_values = [10, 100, 200, 500, 1000, 10000]

# Lista, do której będziemy zapisywać predykcje i błędy dla różnych wartości m
predictions = {}

# Dla każdej wartości m wykonujemy regresję liniową i obliczamy predykcje oraz błąd na zbiorze testowym
for m in m_values:
    # Wybieramy m pierwszych filmów
    first_movies1 = X_train.columns[:m]
    # Tworzymy X zawierające tylko m pierwszych filmów
    X_train_subset = X_train[first_movies1]
    X_test_subset = X_test[first_movies1]
    
    # Tworzymy model regresji liniowej
    model1 = LinearRegression()
    # Dopasowujemy model do danych treningowych
    model1.fit(X_train_subset, y_train)
    # Przewidujemy oceny 'Toy Story' dla zbioru testowego
    y_pred1 = model1.predict(X_test_subset)
    # Zapisujemy predykcje dla wartości m
    predictions[m] = y_pred1

    
for m, y_pred in predictions.items():
    print("Predykcje dla m =", m, ":")
    # Użyj zip(), aby połączyć prawdziwe oceny (y_test) z predykcjami (y_pred)
    for true_rating, predicted_rating in zip(y_test, y_pred):
        print(f"{true_rating}  |  {predicted_rating}")
````

Napisz system rekomendacji filmów który będzie wykorzystywał dane

Interesować, będą nas głównie dwa pliki movies.csv oraz ratings.csv. W pliku ratings.csv mamy
właściwie wszystkie dane, które będą nam potrzebne aby stworzyć macierz oceny. Ponieważ

Otrzymane reprezentacje wyświetl w postaci nazw filmów, korzystając z movies.csv. Uwaga: w rzeczywistych
danych otrzymamy dużą liczbę zer, nawet całe kolumny, wtedy dostaniemy wartości NaN przy dzieleniu!
Rozwiąż ten problem wykorzystując np.nan_to_num(...).



````python
import numpy as np
import pandas as pd

# Wczytanie danych z plików
ratings_df = pd.read_csv("ratings_small.csv")
movies_df = pd.read_csv("movies.csv")

# Wybór tylko tych wierszy, gdzie movieId < 10000
ratings_df = ratings_df[ratings_df['movieId'] < 10000]

# Tworzenie macierzy ocen
ratings_table = ratings_df.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Tworzenie przekształconej do tabeli macierzy ocen 
ratings_matrix = ratings_df.pivot(index='userId', columns='movieId', values='rating').fillna(0).to_numpy()
#print(ratings_matrix)

my_ratings = np.zeros((ratings_matrix.shape[1], 1))

my_ratings[ratings_table.columns.get_loc(2571)] = 5      # Matrix
my_ratings[ratings_table.columns.get_loc(32)] = 4        # Twelve Monkeys
my_ratings[ratings_table.columns.get_loc(260)] = 5       # Star Wars IV
my_ratings[ratings_table.columns.get_loc(1097)] = 4      # E.T. the Extra-Terrestrial

#print(my_ratings.shape)

# Obliczenie normy wektorów kolumnowych macierzy
column_norms = np.linalg.norm(ratings_matrix, axis=0)

# Stworzenie znormalizowanej macierzy
normalized_matrix = np.nan_to_num(ratings_matrix / column_norms)
print("normalized matrix ",normalized_matrix.shape)

# Normalizacja wektora ocen konkretnego użytkownika
my_ratings_norm = np.nan_to_num(my_ratings / np.linalg.norm(my_ratings))

# Obliczenie podobieństwa kosinusowego między użytkownikami a użytkownikiem o znormalizowanych preferencjach -> wektor
user_similarity_scores = np.dot(normalized_matrix, my_ratings_norm)

# Znormalizowanie wektora podobieństw miedzy użytkownikami -> wektor
normalized_user_similarity_scores = np.nan_to_num(user_similarity_scores / np.linalg.norm(user_similarity_scores))

# Obliczenie iloczynu skalarnego miedzy naszym profilem a kazda kolumna macierzy
recommendations = np.dot(normalized_matrix.T, normalized_user_similarity_scores)
#print("recomendations: ", recommendations.flatten())

"""W  `recommendations` indeks $i$ odpowiada $i$-temu filmowi w ratings_matrix (to nie *movieId* !). Sortujemy tablicę malejąco względem podobieństwa, ale tak by zachować informację o indeksach, stąd `np.argsort()`. Indeksy wykorzystujemy do pozyskania faktycznych *movieId* z ratings_table.
"""

matrix_indices_sorted = np.argsort(recommendations.flatten())[::-1]

recommendations_sorted = recommendations[matrix_indices_sorted].flatten()
movie_indices = ratings_table.columns[matrix_indices_sorted]

# Wyświetlenie rekomendacji z nazwami filmów
for i in range(len(recommendations_sorted)):
    movie_id = movie_indices[i]
    movie_title = movies_df[movies_df['movieId'] == movie_id]['title'].values[0]
    print(f"({recommendations_sorted[i]}, {movie_id}, {movie_title})")


````
